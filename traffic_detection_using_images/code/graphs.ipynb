```python
import matplotlib.pyplot as plt

# Example data
models = {
    'YOLOv8': {
        'versions': ['v8n', 'v8s'],
        'fps in GPU': [89.0, 82.64],
        'map': [93.3, 94.9],
        'color': 'blue'
    },
    'YOLOv9': {
        'versions': ['gelan-c', 'yolov9-c'],
        'fps in GPU': [32.89, 16.77],
        'map': [95.5, 95.5],
        'color': 'green'
    },
    'YOLOv10': {
        'versions': ['v10n', 'v10s'],
        'fps in GPU': [67.56, 67.56],
        'map': [92.5, 94.6],
        'color': 'red'
    }
}

# Create the plot
plt.figure(figsize=(10, 6))
for model, data in models.items():
    plt.plot(data['fps in GPU'], data['map'], marker='o', linestyle='-', color=data['color'], label=model)
    for version, fps, map_ in zip(data['versions'], data['fps in GPU'], data['map']):
        plt.text(fps, map_, version, fontsize=12, color=data['color'], ha='right')

# Titles and labels
plt.title('Models trained with 640 image resolution [V100]')
plt.xlabel('FPS in GPU')
plt.ylabel('mAP')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
classes = ['Bus', 'Car', 'Motorbike', 'Bicycle', 'Person']
models = ['yolov8-n', 'yolov8-s', 'yolov9-c', 'gelan-c', 'yolov10-n', 'yolov10-s']
mAP_values = [
    [97.5, 96.8, 89.4, 94.3, 88.3],  # Model 1
    [98.2, 96.8, 92.3, 95.0, 92.4]    ,  # Model 2
    [97.7, 97.1, 93.2, 96.0, 93.3 ],  # Model 3
    [98.0, 96.8, 93.4, 95.9, 93.4],  # Model 4
    [97.3, 96.5, 88.6, 93.1, 87.1],  # Model 5
    [98.5, 96.7, 91.0, 95.7, 91.3]   # Model 6
]

# Number of models
n_models = len(models)

# Setting up the bar width
bar_width = 0.1

# Set position of bar on X axis
r = [np.arange(len(classes))]
for i in range(1, n_models):
    r.append([x + bar_width for x in r[i-1]])

# Create the bar plot
for i in range(n_models):
    plt.bar(r[i], mAP_values[i], color=np.random.rand(3,), width=bar_width, edgecolor='grey', label=models[i])

# Adding Xticks
plt.xlabel('Classes', fontweight='bold')
plt.ylabel('mAP', fontweight='bold')
plt.xticks([r + bar_width * (n_models/2 - 0.5) for r in range(len(classes))], classes)

# Adjust the layout to provide space for the legend
plt.tight_layout(rect=[0, 0, 0.85, 1])  # Adjust the rect parameter to add space on the right

# Add legend to the top-right corner outside the plot
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```
