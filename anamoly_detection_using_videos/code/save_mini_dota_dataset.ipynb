```python
# Rename copied .z01, z.02, z.03, z.04, z.05 and .zip files into their original names

import os

directory_path = '/content/dataset'

# List all files in the directory
files = os.listdir(directory_path)
for file_name in files:
    if file_name.endswith('.zip') or file_name.endswith('.json'):
        pass
    elif os.path.isdir(os.path.join(directory_path, file_name)):
        pass
    else:
        split_name = file_name.split('-')
        print(split_name)
        ext = split_name[-1].split('.')[-1]
        file_new_name = split_name[0]+ '.' + ext
        os.rename(os.path.join(directory_path, file_name), os.path.join(directory_path, file_new_name))
        #print(file_new_name)
```

```python
# Create mini-dataset

import json

json_file_path = '/content/mini_dota_dataset/data/metadata_val.json'
mini_dataset_size = 21

new_json = {}
with open(json_file_path, 'r') as file:
    data = json.load(file)

frame_names = list(data.keys())[:mini_dataset_size]
```

```python
# Copy 20 video frames and corresponding annotations to DOTA mini-dataset

ann_source_path = '/content/dataset/annotations/annotations'
frames_source_path = '/content/dataset/frames'
ann_dest_path = '/content/mini_dota_dataset/data/annotations'
frames_dest_path = '/content/mini_dota_dataset/data/frames'

# mini_dataset_size = 20
source_frames = os.listdir(frames_source_path)

for file_name in frame_names:
    if file_name in source_frames:
      if f"{file_name}.json" in os.listdir(ann_source_path):
        shutil.copy(os.path.join(ann_source_path, f"{file_name}.json"), ann_dest_path)
        shutil.copytree(os.path.join(frames_source_path, file_name.split('.')[0]), os.path.join(frames_dest_path, file_name.split('.')[0]))
```

```python
# quick validation

print(f"copied {len(os.listdir(frames_dest_path))} frames files to mini dataset folder")
print(f"copied {len(os.listdir(ann_dest_path))} annotation files to mini dataset folder")
```

```python
# rewrite meta data file as per small dataset

import json

json_file_path = '/content/mini_dota_dataset/data/metadata_val.json'
frames_path = '/content/mini_dota_dataset/data/frames'
frame_names = os.listdir(frames_path)

new_json = {}
with open(json_file_path, 'r') as file:
    data = json.load(file)

    for key in data.keys():
        if key in frame_names:
            new_json[key] = data[key]

with open(json_file_path, 'w') as file:
    json.dump(new_json, file)
```

```python
# quick validation

with open('/content/mini_dota_dataset/data/metadata_val.json', 'r') as file:
    data = json.load(file)

print(f'metadata_val.json file is overwritten and has {len(data.keys())} frame details')
```
